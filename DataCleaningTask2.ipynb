### Clean the data..(missing values,outliers,type casting,duplicate,variance,...etc)

### missing values

```python
data.isna().sum()
```

```python
max_id = data['Customer ID'].max()   # customer id
data['Customer ID'].fillna(max_id + 1, inplace=True)
```

```python
data["Age"].fillna(data["Age"].mode()[0],inplace = True)             # age

data["Gender"].fillna(data["Gender"].mode()[0],inplace = True)      # gender

data["Item Purchased"].fillna(data["Item Purchased"].mode()[0],inplace = True)  #Item Purchased

data["Category"].fillna(data["Category"].mode()[0],inplace = True) # category

data["Location"].fillna(data["Location"].mode()[0],inplace = True) # location

data["Color"].fillna(data["Color"].mode()[0],inplace = True) # color

data["Season"].fillna(data["Season"].mode()[0],inplace = True)  # season

data["Review Rating"].fillna(data["Review Rating"].mode()[0],inplace = True) # review rating

data["Subscription Status"].fillna(data["Subscription Status"].mode()[0],inplace = True) # subscription status

data["Discount Applied"].fillna(data["Discount Applied"].mode()[0],inplace = True) # discount applied

data["Promo Code Used"].fillna(data["Promo Code Used"].mode()[0],inplace = True)  # promo code used

data["Payment Method"].fillna(data["Payment Method"].mode()[0],inplace = True) # payment metod

data["Frequency of Purchases"].fillna(data["Frequency of Purchases"].mode()[0],inplace = True) # Frequency of Purchases
```

### Outliers

```python
data[(data["Age"] < 0) | (data["Age"] > 100)] # age

data["Age"].fillna(data["Age"].mode()[0],inplace = True)

condition = (data["Review Rating"] < 0) | (data["Review Rating"] > 5)   # review rating

data["Review Rating"] = np.where(condition,3.5,data["Review Rating"])
```

```python
data
```

### type casting

```python
data["Age"] = data["Age"].astype("float32") # convert float64 to float32

data["Purchase Amount (USD)"] = data["Purchase Amount (USD)"].astype("int16") # convert int64 to int 16
  
data["Review Rating"] = data["Review Rating"].astype("float32") # convert float64 to float32
```

### Scalling

```python
data["Age"].mean(),data["Age"].median()  # age
```

```python
def standard(x):    # age
    n = (x - data["Age"].mean()) / data["Age"].std()
    return n
data["Age"] = data["Age"].apply(standard)
```

```python
data["Purchase Amount (USD)"].mean(),data["Purchase Amount (USD)"].median()  # Purchase Amount (USD)
```

```python
def standard(x):    # Purchase Amount (USD)
    n = (x - data["Purchase Amount (USD)"].mean()) / data["Purchase Amount (USD)"].std()
    return n
data["Purchase Amount (USD)"] = data["Purchase Amount (USD)"].apply(standard)
```

```python
data["Review Rating"].mean(),data["Review Rating"].median()  # review rating
```

```python
def standard(x):    # review rating
    n = (x - data["Review Rating"].mean()) / data["Review Rating"].std()
    return n
data["Review Rating"] = data["Review Rating"].apply(standard)
```
