### using domain knowledge

```python
condition = (df["SepalLengthCm"] > 8.0) | (df["SepalLengthCm"] < 0)  # sepallengthcm

df["SepalLengthCm"] = np.where(condition,6.0,df["SepalLengthCm"])
```

```python
condition = (df["SepalWidthCm"] > 8.0) | (df["SepalWidthCm"] < 0) # sepalwidthcm

df["SepalWidthCm"] = np.where(condition,4.0,df["SepalWidthCm"])
```

```python
condition = (df["PetalLengthCm"] > 8.0) | (df["PetalLengthCm"] < 0)  # sepallengthcm

df["PetalLengthCm"] = np.where(condition,7.0,df["PetalLengthCm"])
```

```python
condition = (df["PetalWidthCm"] > 8.0) | (df["PetalWidthCm"] < 0) # petalwidthcm

df["PetalWidthCm"] = np.where(condition,6.0,df["PetalWidthCm"])
```
