### 1.clean the data(outlier,missing values,type conversion)

### sepal width

```python
Iris = np.genfromtxt(r"C:\Users\LENOVO\Downloads\Iris_clean_-_Copy.csv",dtype = "float64",delimiter = ",")
Iris
```

```python
width = Iris[1:,1]
width
```

```python
Iris = np.where(width == -1,np.nan,width) #nan values
Iris
```

```python
np.nanmedian(width)
```

```python
Iris = np.where(np.isnan(width),np.nanmedian(width),width)
Iris
```

```python
con = (Iris > 5) | (Iris < 2) # outlier
```

```python
sepal_width = np.where(con,3,Iris)
sepal_width
```

```python
sepal_width = np.array(sepal_width,dtype = "float16")
sepal_width
```

### sepal length

```python
length = Iris[1:,0]
length
```

```python
np.nanmedian(length)
```

```python
a = np.where(np.isnan(length),np.nanmedian(length),length)  # replace nan with nanmedian
a
```

```python
con = a > 8  # outlier
```

```python
sepallength = np.where(con,5,a)
sepallength
```

```python
sepal_length = np.array(sepallength,dtype = "float16")
sepal_length
```
