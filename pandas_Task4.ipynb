### 3.Filling  method

```python
ethnic = pd.read_excel(r"C:\Users\LENOVO\Downloads\ethnic_diversity.xlsx")
ethnic
```

```python
ethnic.isna().sum()
```

```python
ethnic["Position"].fillna(method="ffill",inplace=True)   # position

ethnic["State"].fillna(method="ffill",inplace=True)   # state

ethnic["Zip"].fillna(method="ffill",inplace=True)   # zip

ethnic["Gender"].fillna(method="ffill",inplace=True)   # Gender

ethnic["MaritalDesc"].fillna(method="ffill",inplace=True)   # maritaldesc

ethnic["CitizenDesc"].fillna(method="ffill",inplace=True)   # maritaldesc

ethnic["Department"].fillna(method="ffill",inplace=True)   # department

ethnic["Salaries"].fillna(method="ffill",inplace=True)   # salaries

ethnic["Race"].fillna(method="ffill",inplace=True)   # race
```
